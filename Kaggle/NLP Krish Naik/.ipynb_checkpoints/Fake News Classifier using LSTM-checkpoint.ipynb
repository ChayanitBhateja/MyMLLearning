```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/fakeNewsTrainData.csv')
df.head()
```

```python
 df = df.dropna()
```

```python
features = df.drop('label', axis = 1)
```

```python
label = df['label']
```

```python
label.shape
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size = 5000
```

# One Hot Representation..

```python
messages = features.copy().reset_index().drop('index', axis = 1)

messages.head(10)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
corpus = []
ps = PorterStemmer()
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
    
corpus
```

```python
onehot_repr = [one_hot(words, voc_size) for words in corpus]
onehot_repr
```

# Embedding Representation..

```python
sent_length = 20
embedding_vector_features = 40
embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen = sent_length)
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer ='adam',metrics='mse')
print(model.summary())
```

```python
embedded_docs[0]
```

```python
import numpy as np 
featuresFinal = np.array(embedded_docs)
labelFinal = np.array(label)
```

```python
featuresFinal.shape, labelFinal.shape
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(featuresFinal, labelFinal, test_size = 0.33)
xTrain.shape,yTrain.shape
```

```python
model.fit(xTrain, yTrain, validation_data=(xTest, yTest), epochs=10, batch_size = 64)
```

```python
y_pred = (model.predict(xTest)>0.5).astype('int32')
y_pred
```

```python
yTestArray = np.array(yTest)
```

```python
yTestArray
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(yTest,y_pred)
```

```python
accuracy_score(yTestArray,y_pred)
```

```python
from tensorflow.keras.layers import Dropout
```

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length = sent_length))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dropout(0.2))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics='accuracy')
```

```python
model.fit(xTrain, yTrain, validation_data=(xTest, yTest), epochs = 10)
```

```python
prediction = (model.predict(xTest) > 0.5).astype('int32')
confusion_matrix(yTest, prediction),accuracy_score(prediction,yTest)
```
