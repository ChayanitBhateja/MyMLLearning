```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('vaccination_tweets.csv')
dataset.head()
```

```python
tweets = np.array(dataset.text)
tweets
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size= 10000
one_hot_repr = [one_hot(words, voc_size) for words in tweets]
print(one_hot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
tweetLength = 25
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen = tweetLength)
# print(embedded_docs)
print(embedded_docs[:,0])
```

```python
dimension = 10
model = Sequential()
model.add(Embedding(voc_size, dimension, input_length = tweetLength))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs)[0][0])
```
